```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_cleaned_dataset1.csv')
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
# Drop day,month & year
df.drop(['day','month','year'],axis =1, inplace = True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## Encoding
df['Classes'] = np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## Independent and Dependent features
#X = df.drop('FWI',axis = 1)
#y = df['FWI']
X=df.iloc[:,df.columns != df.columns[-3]]
y=df.iloc[:,9]
```

```python
X.head()
```

```python
y.head()
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.25, random_state =42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
## Feature Selection Based on Correlation
X_train.corr()
```

```python
## Check for Multicollinearity
plt.figure(figsize = (12,10))
corr = X_train.corr()
sns.heatmap(corr,annot = True)
```

```python
def correlation(dataset, threshold):
    col_corr = set() # Initialize an empty set to store the names of highly correlated columns
    corr_matrix = dataset.corr() # Calculate the correlation matrix for the dataset
    for i in range(len(corr_matrix.columns)): # This iterates over each column in the correlation matrix, starting from the first column.
        for j in range(i): # This loops through each row before the column i. It ensures the function only looks at each pair of columns once (only the upper or lower triangle of the correlation matrix), avoiding duplicate comparisons.
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## threshold--Domain expertise
corr_features=correlation(X_train,0.85)
corr_features
```

```python
## drop features when correlation is more than 0.85 
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

# Feature Scaling or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

# Box Plots To understand Effect Of Standard Scaler

```python
plt.subplots(figsize=(20, 12))
plt.subplot(2, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(2, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
plt.subplot(2, 2, 3)
sns.boxplot(data=X_test)
plt.title('X_test Before Scaling')
plt.subplot(2, 2, 4)
sns.boxplot(data=X_test_scaled)
plt.title('X_test After Scaling')
plt.tight_layout()  # Adjusts spacing between the subplots
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linreg=LinearRegression()
linreg.fit(X_train_scaled,y_train)
y_pred=linreg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(y_test,y_pred)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(y_test,y_pred)
```

# Cross Validation Lasso

# Ridge Regression model

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, r2_score
ridge = Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred = ridge.predict(X_test_scaled)
mae = mean_absolute_error(y_pred,y_test)
score = r2_score(y_pred,y_test)
print('MAE:', mae)
print('R2:', score)
plt.scatter(y_pred,y_test)
```

```python
from sklearn.linear_model import RidgeCV
ridgeCV = RidgeCV(cv = 5)
ridgeCV.fit(X_train_scaled,y_train)
y_pred = ridgeCV.predict(X_test_scaled)
mae = mean_absolute_error(y_pred,y_test)
score = r2_score(y_pred,y_test)
print('MAE:',mae)
print('R2:',score)
plt.scatter(y_pred,y_test)
```
